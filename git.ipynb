```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,2))
ax = plt.subplot(2,1,1)
ax.invert_yaxis()
plt.barh(1, 5, label='actual work')
stops = [0.5, 1.3, 1.9, 2.4, 3.5]
for stop in stops:
    plt.barh(1, 0.2, left=stop, color='red')
plt.barh(1, 0.9, left=4.1, color='red', label='fixing stuff')
plt.axis('off')
plt.legend(loc='upper left', bbox_to_anchor=(0.00,2.2), fontsize=16)

ax = plt.subplot(2,1,2)
ax.invert_yaxis()
plt.barh(1, 5)
plt.barh(1, 0.4, left=0, color='red')
for stop in stops:
    plt.barh(1, 0.025, left=stop, color='red')
plt.axis('off')
plt.savefig('imgs/project_workflow.png', bbox_inches="tight")
plt.show()
```
